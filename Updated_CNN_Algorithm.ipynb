```python
import os
import zipfile
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from matplotlib import pyplot as plt
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Extract images from zip file
zip_path = '/content/drive/MyDrive/Colab Notebooks/Image_project_pattern.zip'
extract_path = '/tmp/Image_project_pattern'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List of image files and labels
image_files_brut = [
    'Early_morning_01.jpg', 'Early_morning_02.jpg', 'Early_morning_03.jpg', 'Early_morning_04.jpg',
    'Early_morning_05.jpg', 'Early_morning_06.jpg', 'Early_morning_07.jpg', 'End_morning_01.jpg', 'End_morning_02.jpg',
    'End_morning_03.jpg', 'End_morning_04.jpg', 'End_morning_05.jpg', 'End_morning_06.jpg',
    'End_Morning_07.jpg', 'Early_evening_01.jpg', 'Early_evening_02.jpg', 'Early_evening_03.jpg', 'Early_evening_04.jpg',
    'Early_evening_05.jpg', 'Early_evening_06.jpg', 'Early_evening_07.jpg', 'late_evening_01.jpg', 'late_evening_02.jpg',
    'late_evening_03.jpg', 'late_evening_04.jpg', 'late_evening_05.jpg', 'late_evening_06.jpg', 'late_evening_07.jpg'
]

labels = [
    'Early_morning', 'Early_morning', 'Early_morning', 'Early_morning',
    'Early_morning', 'Early_morning', 'Early_morning', 'Late_morning', 'Late_morning',
    'Late_morning', 'Late_morning', 'Late_morning', 'Late_morning',
    'Late_morning', 'early_evening', 'early_evening', 'early_evening', 'early_evening',
    'early_evening', 'early_evening', 'early_evening', 'Late_evening', 'Late_evening',
    'Late_evening', 'Late_evening', 'Late_evening', 'Late_evening', 'Late_evening'
]

# Get the actual filenames in the directory
actual_filenames = {f.lower(): f for f in os.listdir(extract_path)}

# Filter image_files_brut and Labels to match the actual files in the directory
filtered_image_files_brut = [actual_filenames[img.lower()] for img in image_files_brut if img.lower() in actual_filenames]
filtered_labels = [label for img, label in zip(image_files_brut, labels) if img.lower() in actual_filenames]

# Ensure consistent length
assert len(filtered_image_files_brut) == len(filtered_labels), "Mismatch between the number of images and labels after filtering."

# Preprocess and crop the images
def preprocess_image(im_path, top_reduce=500, bottom_reduce=500, left_reduce=500, right_reduce=500):
    image = plt.imread(im_path) / 255.0  # Normalize the image
    height, width, _ = image.shape
    top_boundary = top_reduce
    bottom_boundary = height - bottom_reduce
    left_boundary = left_reduce
    right_boundary = width - right_reduce
    if bottom_boundary <= top_boundary or right_boundary <= left_boundary:
        raise ValueError("Cropping dimensions exceed image dimensions.")
    return image[top_boundary:bottom_boundary, left_boundary:right_boundary]

image_files = [preprocess_image(os.path.join(extract_path, im)) for im in filtered_image_files_brut]

# Convert categorical labels to numerical indices
label_to_index = {label: idx for idx, label in enumerate(set(filtered_labels))}
indexed_labels = [label_to_index[label] for label in filtered_labels]

# Split the data into training and testing using StratifiedShuffleSplit
shuffle = StratifiedShuffleSplit(n_splits=1, test_size=0.14, random_state=42)
train_index, test_index = next(shuffle.split(image_files, indexed_labels))
X_train = [image_files[i] for i in train_index]
X_test = [image_files[i] for i in test_index]
y_train = [indexed_labels[i] for i in train_index]
y_test = [indexed_labels[i] for i in test_index]

# Define random_batch function
def random_batch(image, batch_width, batch_height, n_batches):
    height, width, _ = image.shape
    cropped_images = []
    for _ in range(n_batches):
        max_x = width - batch_width
        max_y = height - batch_height
        if max_x < 0 or max_y < 0:
            raise ValueError("Batch dimensions exceed image dimensions.")
        x = random.randint(0, max_x)
        y = random.randint(0, max_y)
        cropped_image = image[y:y+batch_height, x:x+batch_width]
        cropped_images.append(cropped_image)
    return cropped_images

# Define batch parameters
batch_width = 100
batch_height = 100
n_batches = 5

# Create batches from each image
samples_train = [random_batch(img, batch_width, batch_height, n_batches) for img in X_train]
samples_test = [random_batch(img, batch_width, batch_height, n_batches) for img in X_test]

# Flatten the list of batches
imgs_train = [img for batch in samples_train for img in batch]
imgs_test = [img for batch in samples_test for img in batch]

# Repeat labels to match the number of batches
labels_train = np.repeat(y_train, n_batches)
labels_test = np.repeat(y_test, n_batches)

# Convert labels to categorical
labels_categorical_train = to_categorical(labels_train)
labels_categorical_test = to_categorical(labels_test)

# Convert lists to numpy arrays
X_train = np.array(imgs_train)
X_test = np.array(imgs_test)

# Enhanced Data Augmentation
augmentation = ImageDataGenerator(
    rotation_range=50,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

# Define a function to create a new model instance
def create_model():
    base_model = tf.keras.applications.MobileNetV2(weights='imagenet', include_top=False, input_shape=(batch_height, batch_width, 3))
    base_model.trainable = False  # Freeze the base model
    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(len(set(indexed_labels)), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

lr_scheduler = callbacks.LearningRateScheduler(scheduler)

# Cross-Validation
kf = StratifiedKFold(n_splits=7)
train_accuracies = []
val_accuracies = []
train_losses = []
val_losses = []
histories = []
for fold_no, (train_index, val_index) in enumerate(kf.split(X_train, labels_train), start=1):
    print(f'Training fold {fold_no} ...')

    X_train_fold = np.array([X_train[i] for i in train_index])
    y_train_fold = np.array([labels_train[i] for i in train_index])
    X_val_fold = np.array([X_train[i] for i in val_index])
    y_val_fold = np.array([labels_train[i] for i in val_index])

    model = create_model()

    history = model.fit(
        augmentation.flow(X_train_fold, to_categorical(y_train_fold), batch_size=16),
        epochs=50,
        validation_data=(X_val_fold, to_categorical(y_val_fold)),
        callbacks=[lr_scheduler]
    )

    train_accuracies.append(history.history['accuracy'])
    val_accuracies.append(history.history['val_accuracy'])
    train_losses.append(history.history['loss'])
    val_losses.append(history.history['val_loss'])
    histories.append(history)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, labels_categorical_test)
print(f"Test accuracy: {test_acc*100:.2f}%")

# Predict and display classification results
predictions = model.predict(X_test)

# Replace cluster labels with the actual labels
predicted_labels = pd.Series(np.argmax(predictions, axis=1)).replace({
    0: 'Early_morning',
    1: 'Late_morning',
    2: 'early_evening',
    3: 'Late_evening'
})

# Display classification results
def display_classification_results(X, y, predictions):
    plt.figure(figsize=(10, 10))
    for i in range(len(predictions)):
        ax = plt.subplot(4, 5, i + 1)
        plt.imshow(X[i])
        pred_label = pd.Series(np.argmax(predictions, axis=1)).replace({
            0: 'Early_morning',
            1: 'Late_morning',
            2: 'early_evening',
            3: 'Late_evening'
        })[i]
        true_label = pd.Series(np.argmax(y, axis=1)).replace({
            0: 'Early_morning',
            1: 'Late_morning',
            2: 'early_evening',
            3: 'Late_evening'
        })[i]
        plt.title(f"Pred: {pred_label}, True: {true_label}")
        plt.axis("off")
        if i == 4:  # Display first 20 images
            break
    plt.show()

display_classification_results(X_test, labels_categorical_test, predictions)

# Plot training accuracy
plt.figure(figsize=(12, 5))
for train_acc in train_accuracies:
    plt.plot(train_acc)
plt.title('Training Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(train_accuracies))], loc='upper left')
plt.show()

# Plot validation accuracy
plt.figure(figsize=(12, 5))
for val_acc in val_accuracies:
    plt.plot(val_acc)
plt.title('Validation Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(val_accuracies))], loc='upper left')
plt.show()

# Plot training loss
plt.figure(figsize=(12, 5))
for train_loss in train_losses:
    plt.plot(train_loss)
plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(train_losses))], loc='upper left')
plt.show()

# Plot validation loss
plt.figure(figsize=(12, 5))
for val_loss in val_losses:
    plt.plot(val_loss)
plt.title('Validation Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(val_losses))], loc='upper left')
plt.show()
```

```python
import os
import zipfile
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from matplotlib import pyplot as plt
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Extract images from zip file
zip_path = '/content/drive/MyDrive/Colab Notebooks/Image_project_pattern.zip'
extract_path = '/tmp/Image_project_pattern'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List of image files and labels
image_files_brut = [
    'Early_morning_01.jpg', 'Early_morning_02.jpg', 'Early_morning_03.jpg', 'Early_morning_04.jpg',
    'Early_morning_05.jpg', 'Early_morning_06.jpg', 'Early_morning_07.jpg', 'End_morning_01.jpg', 'End_morning_02.jpg',
    'End_morning_03.jpg', 'End_morning_04.jpg', 'End_morning_05.jpg', 'End_morning_06.jpg',
    'End_Morning_07.jpg', 'Early_evening_01.jpg', 'Early_evening_02.jpg', 'Early_evening_03.jpg', 'Early_evening_04.jpg',
    'Early_evening_05.jpg', 'Early_evening_06.jpg', 'Early_evening_07.jpg', 'late_evening_01.jpg', 'late_evening_02.jpg',
    'late_evening_03.jpg', 'late_evening_04.jpg', 'late_evening_05.jpg', 'late_evening_06.jpg', 'late_evening_07.jpg'
]

labels = [
    'Early_morning', 'Early_morning', 'Early_morning', 'Early_morning',
    'Early_morning', 'Early_morning', 'Early_morning', 'Late_morning', 'Late_morning',
    'Late_morning', 'Late_morning', 'Late_morning', 'Late_morning',
    'Late_morning', 'early_evening', 'early_evening', 'early_evening', 'early_evening',
    'early_evening', 'early_evening', 'early_evening', 'Late_evening', 'Late_evening',
    'Late_evening', 'Late_evening', 'Late_evening', 'Late_evening', 'Late_evening'
]

# Get the actual filenames in the directory
actual_filenames = {f.lower(): f for f in os.listdir(extract_path)}

# Filter image_files_brut and Labels to match the actual files in the directory
filtered_image_files_brut = [actual_filenames[img.lower()] for img in image_files_brut if img.lower() in actual_filenames]
filtered_labels = [label for img, label in zip(image_files_brut, labels) if img.lower() in actual_filenames]

# Ensure consistent length
assert len(filtered_image_files_brut) == len(filtered_labels), "Mismatch between the number of images and labels after filtering."

# Preprocess and crop the images
def preprocess_image(im_path, top_reduce=500, bottom_reduce=500, left_reduce=500, right_reduce=500):
    image = plt.imread(im_path) / 255.0  # Normalize the image
    height, width, _ = image.shape
    top_boundary = top_reduce
    bottom_boundary = height - bottom_reduce
    left_boundary = left_reduce
    right_boundary = width - right_reduce
    if bottom_boundary <= top_boundary or right_boundary <= left_boundary:
        raise ValueError("Cropping dimensions exceed image dimensions.")
    return image[top_boundary:bottom_boundary, left_boundary:right_boundary]

image_files = [preprocess_image(os.path.join(extract_path, im)) for im in filtered_image_files_brut]

# Convert categorical labels to numerical indices
label_to_index = {label: idx for idx, label in enumerate(set(filtered_labels))}
indexed_labels = [label_to_index[label] for label in filtered_labels]

# Split the data into training and testing using StratifiedShuffleSplit
shuffle = StratifiedShuffleSplit(n_splits=1, test_size=0.14, random_state=42)
train_index, test_index = next(shuffle.split(image_files, indexed_labels))
X_train = [image_files[i] for i in train_index]
X_test = [image_files[i] for i in test_index]
y_train = [indexed_labels[i] for i in train_index]
y_test = [indexed_labels[i] for i in test_index]

# Define random_batch function
def random_batch(image, batch_width, batch_height, n_batches):
    height, width, _ = image.shape
    cropped_images = []
    for _ in range(n_batches):
        max_x = width - batch_width
        max_y = height - batch_height
        if max_x < 0 or max_y < 0:
            raise ValueError("Batch dimensions exceed image dimensions.")
        x = random.randint(0, max_x)
        y = random.randint(0, max_y)
        cropped_image = image[y:y+batch_height, x:x+batch_width]
        cropped_images.append(cropped_image)
    return cropped_images

# Define batch parameters
batch_width = 100
batch_height = 100
n_batches = 5

# Create batches from each image
samples_train = [random_batch(img, batch_width, batch_height, n_batches) for img in X_train]
samples_test = [random_batch(img, batch_width, batch_height, n_batches) for img in X_test]

# Flatten the list of batches
imgs_train = [img for batch in samples_train for img in batch]
imgs_test = [img for batch in samples_test for img in batch]

# Repeat labels to match the number of batches
labels_train = np.repeat(y_train, n_batches)
labels_test = np.repeat(y_test, n_batches)

# Convert labels to categorical
labels_categorical_train = to_categorical(labels_train)
labels_categorical_test = to_categorical(labels_test)

# Convert lists to numpy arrays
X_train = np.array(imgs_train)
X_test = np.array(imgs_test)

# Enhanced Data Augmentation
augmentation = ImageDataGenerator(
    rotation_range=50,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

def create_custom_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(batch_height, batch_width, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(len(set(indexed_labels)), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model


# Learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

lr_scheduler = callbacks.LearningRateScheduler(scheduler)

# Cross-Validation
kf = StratifiedKFold(n_splits=7)
train_accuracies = []
val_accuracies = []
train_losses = []
val_losses = []
histories = []
for fold_no, (train_index, val_index) in enumerate(kf.split(X_train, labels_train), start=1):
    print(f'Training fold {fold_no} ...')

    X_train_fold = np.array([X_train[i] for i in train_index])
    y_train_fold = np.array([labels_train[i] for i in train_index])
    X_val_fold = np.array([X_train[i] for i in val_index])
    y_val_fold = np.array([labels_train[i] for i in val_index])

    model = create_custom_model()  # Use custom model here

    history = model.fit(
        augmentation.flow(X_train_fold, to_categorical(y_train_fold), batch_size=16),
        epochs=50,
        validation_data=(X_val_fold, to_categorical(y_val_fold)),
        callbacks=[lr_scheduler]
    )

    train_accuracies.append(history.history['accuracy'])
    val_accuracies.append(history.history['val_accuracy'])
    train_losses.append(history.history['loss'])
    val_losses.append(history.history['val_loss'])
    histories.append(history)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, labels_categorical_test)
print(f"Test accuracy: {test_acc*100:.2f}%")

# Predict and display classification results
predictions = model.predict(X_test)

# Replace cluster labels with the actual labels
predicted_labels = pd.Series(np.argmax(predictions, axis=1)).replace({
    0: 'Early_morning',
    1: 'Late_morning',
    2: 'early_evening',
    3: 'Late_evening'
})

# Display classification results
def display_classification_results(X, y, predictions):
    plt.figure(figsize=(10, 10))
    for i in range(len(predictions)):
        ax = plt.subplot(4, 5, i + 1)
        plt.imshow(X[i])
        pred_label = pd.Series(np.argmax(predictions, axis=1)).replace({
            0: 'Early_morning',
            1: 'Late_morning',
            2: 'early_evening',
            3: 'Late_evening'
        })[i]
        true_label = pd.Series(np.argmax(y, axis=1)).replace({
            0: 'Early_morning',
            1: 'Late_morning',
            2: 'early_evening',
            3: 'Late_evening'
        })[i]
        plt.title(f"Pred: {pred_label}, True: {true_label}")
        plt.axis("off")
        if i == 4:  # Display first 20 images
            break
    plt.show()

display_classification_results(X_test, labels_categorical_test, predictions)

# Plot training accuracy
plt.figure(figsize=(12, 5))
for train_acc in train_accuracies:
    plt.plot(train_acc)
plt.title('Training Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(train_accuracies))], loc='upper left')
plt.show()

# Plot validation accuracy
plt.figure(figsize=(12, 5))
for val_acc in val_accuracies:
    plt.plot(val_acc)
plt.title('Validation Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(val_accuracies))], loc='upper left')
plt.show()

# Plot training loss
plt.figure(figsize=(12, 5))
for train_loss in train_losses:
    plt.plot(train_loss)
plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(train_losses))], loc='upper left')
plt.show()

# Plot validation loss
plt.figure(figsize=(12, 5))
for val_loss in val_losses:
    plt.plot(val_loss)
plt.title('Validation Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend([f'Fold {i+1}' for i in range(len(val_losses))], loc='upper left')
plt.show()
```
